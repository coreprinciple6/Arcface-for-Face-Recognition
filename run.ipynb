```python
#import libraries
from torchvision import transforms as trans
from training.config import get_config
from training.Learner import face_learner
```

```python
#set configurations
epoch = 2
lr = 1e-3
batch_size = 10
conf = get_config()
```

```python
#define hyperparams
conf.lr = lr
conf.batch_size = batch_size
learner = face_learner(conf)
```

```python
#call training functionr
acc, roc_curve_tensor = learner.train(conf, epoch)
```

```python
print('Accuracy:', acc*100,'%')
```
